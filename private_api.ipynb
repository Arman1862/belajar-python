<p style="text-align:center;">Environment Variables</p>

os is a powerful module. With its getenv function, we can retrieve a single environment variable at a time if we provide the key.

If we query a non-existent environment variable using os.getenv, the result will be None.

```python
import os

api_key = os.getenv("MY_API_KEY")
print(api_key)
```

Environment variables should always be securely stored to protect sensitive information.

You should never hardcode your API keys or passwords; the following example is for educational purposes to demonstrate their structure.

If you save your environment variables in a .env file, you can use the dotenv library to load them into your environment.

Once loaded, you can access them using os.getenv in your code.

```python
import dotenv
import os

dotenv.load_dotenv()
api_key = os.getenv("API_KEY")
print(api_key)
```

Remember os.environ is a dictionary. You can use it like any other dictionary. While os.getenv is a method that returns a value.

```python
import os

api_key = os.environ["API_KEY"]
secret_key = os.getenv("SECRET_KEY")
print(api_key)
print(secret_key)
```

If we query a non-existent environment variable using os.environ, it will raise a KeyError because the variable does not exist.

```python
import os

api_key = os.environ["NON_EXISTENT_KEY"]
print(api_key)
```

You can set a default value for an environment variable in code, but never use a private key for this to ensure security.

```python
import os

api_key = os.getenv("NON_EXISTENT_KEY", "default_value")
print(api_key)
```

When working with the os.environ dictionary, we can use the get dictionary method to also provide a default value if needed.

```python
import os

api_key = os.environ.get("NON_EXISTENT_KEY", "default_value")
print(api_key)
```

<p style="text-align:center;">Customming API Calls</p>

APIs often require authentication details to be sent through API headers. We can specify these headers while making the API request.

Headers in an API request are typically defined as a dictionary, with each key representing the header name.

```python
import requests

headers = {"api-key": "MY_SECRET_KEY"}

requests.get("http://mimo.org/api/exclusive_guest_book", headers=headers)
```

We can directly pass the dictionary to the headers parameter without saving it in a variable first.

However, for readability, it's best to define headers separately, particularly if they are complex or reused across multiple requests.

```python
import requests

requests.get("http://mimo.org/api/exclusive_guest_book", headers={"api-key": "MY_SECRET_KEY"})
```

```python
import requests

headers = {"api-key": "MY_SECRET_KEY"}
body = {"my-message": "I was here!"}

response = requests.post("http://mimo.org/api/exclusive_guest_book", headers=headers, json=body)
print(response.content)
```

We can pass the dictionary with the query parameters as params to requests.get.

```python
import requests

params = {
  "popularity": "high",
  "limit": 1
}

response = requests.get("https://mimo.org/api/languages", params=params)
print(response.json())
```

Every response object includes a status code indicating the result of the request.

We access the status code using response.status_code.

```python
import requests

response = requests.get("https://pokedex.mimo.dev/api/pokemon/pikachu")
print(response.status_code)
```

When working with an API, it often helps to print the full response data to see if we receive what we expect.

```python
import requests

response = requests.get("https://pokedex.mimo.dev/api/pokemon/pikachu")
data = response.json()
print(data["name"])
```

Once you've unpacked the array, you can loop through its elements or directly access an item by its index.

```python
import requests

response = requests.get("https://pokedex.mimo.dev/api/pokemon")
data = response.json()
for pokemon in data:
  print(f"Name: {pokemon['name']}, ID: {pokemon['id']}") 
```
