<p style="text-align:center;">Using Return</p>

```python
def say_hello(name):
    say = "Hello " + name
    return say
```

```python
print(say_hello("Mahen"))
```

```python
call_say = say_hello("Arman")
print(call_say)
```

<p style="text-align:center;">Using Parameter</p>

```python
def say_multiple(firstName, secondName, thirdName):
    return "All name " + firstName + " " + secondName + " " + thirdName

do_say = say_multiple("Arman", "Julio", "Rama")
print(do_say)
```
