<p style="text-align:center;">Variabel Scope</p>

```python
# local variabel = dibikin di function gbisa dipake diluar
# global variabel = bisa dipake diluar function
# super global variabel = bisa dipake walau beda file
```

```python
# local

def name_scope():
    name = "Arman"
    return name

print(name_scope())

# access variabel name
print(name) # error karna tidak terdefinisi
```

```python
# global
name = "Pangestu "

def name_global_scope(add):
    return name + add

print(name_global_scope("Ramadhan"))


```

<p style="text-align:center;">Function with Looping</p>

```python
# Looping Using While

def function_notification_while(message):
    notification = 0
    while notification < message:
        notification += 1
        print(f"You have {notification} message unread from {message}")

function_notification_while(2)
```

```python
# Looping Using For

def function_notification_for(counter):
    for i in range(counter):
        print(f"You have seen {i+1} message from {counter}")

function_notification_for(3)
```
