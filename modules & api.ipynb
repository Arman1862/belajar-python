```python
# try and catch digunakan untuk mengetes program yang
# yang kemungkinan bakal error
# except berjalan kalo try gajalan          
# pass untuk except tidak return apa apa
# bisa dibarengin sama else kalo pake try
```

```python
try:
    operation = 2 / 0
except:
    print("Operation not valid")
```

```python
scores = [20, 10]
average = 0
```

```python
try:
    average = sum(scores)/len(scores)
except:
    average = 0

print(average)
```

<p style="text-align:center;">Macam Macam Response Object</p>

```python
# .status_code buat mengecek statusnya(get)

url = "https://arman.my.id"

response = requests.get(url)
print(response.status_code)
```

```python
import requests 
 
url = "https://pokedex.mimo.dev/api/pokemon/pikachu" 
try: 
  response = requests.get(url) 
  response.raise_for_status() 
  print("Data successfully retrieved") 
  pokemon_data = response.json() 
  print(f"Name: {pokemon_data['name']}") 
  print(f"ID: {pokemon_data['id']}") 
except requests.HTTPError as error: 
  print(error)
```
