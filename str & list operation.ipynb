<p style="text-align:center;">Comprehension(Pemahaman) Strings</p>

```python
prices = [10, 38, 40, 58, 62]
halved = []

for price in prices:
  half_price = price/2
  halved.append(half_price)

print(halved)
```

```python
# bisa dalam one line
prices = [10, 38, 40, 58, 62]

halved = [price/2 for price in prices]
print(halved)
```

```python
ages = [15, 20, 19]

can_drive = [age >= 18 for age in ages]

print(can_drive)
```

```python
def halve(num):
    return num/2

halved = [halve(price) for price in prices]

print(halved)
```

```python
#Penambah koma
authors = ["Virginia Woolf", "John Steinbeck"]

def add_comma(name):
  parts = name.split(" ")
  return parts[0] + ", " + parts[1]

authors_update = [add_comma(name) for name in authors]

print(authors_update)
```

<p style="text-align:center;">Filtering With If Statement</p>

```python
scores = [10, 20, 30, 40, 50]

more_20 = [ score for score in scores if score > 20]

print(more_20)
```

```python
websites = ["nytimes.com", "lemonde.fr", "economist.com"]

french = [website for website in websites if website.count(".fr") > 0]

print(french)
```

<p style="text-align:center;">Transaction Analyzer 1</p>

```python
data = [
  (749.17, "Investment Return"),
  (-11.54, "Utilities"),
  (-247.58, "Online Shopping"),
  (981.17, "Investment Return"),
  (-410.65, "Rent"),
  (310.60, "Rent"),
  (563.70, "Gift"),
  (220.79, "Salary"),
  (-49.85, "Car Maintenance"),
  (308.49, "Salary"),
  (-205.55, "Car Maintenance"),
  (870.64, "Salary"),
  (-881.51, "Utilities"),
  (518.14, "Salary"),
  (-264.66, "Groceries")
]

def print_transactions(transactions):
    for transaction in transactions:
        amount, statement = transaction
        print(f"${amount} - {statement}")

print_transactions(data)

def print_summary(transactions):
    # deposits = [transaction[0] for transaction in transactions]
    deposits = [transaction[0] for transaction in transactions if transaction[0] >= 0]
    total_deposits = sum(deposits)
    print(total_deposits)
    withdrawals = [transaction[0] for transaction in transactions if transaction[0] < 0]
    total_withdrawn = sum(withdrawals)
    print(total_withdrawn)
    balance = (total_deposits + total_withdrawn) / 2
    print(balance)

print_summary(data)
```

```python
savings = [120, 300, 50]

def exchange(dollars):
  euros = 0.8 * dollars
  comission = 0.1 * euros
  return euros - comission

euros = [exchange(amount) for amount in savings]

print(euros)
```

<p style="text-align:center;">Negative Indexing and Deletion</p>

```python
#The del keyword allows us to delete objects, or items within a data structure.
winners = ["John", "Helen", "Sigmund", "Olaf"]
del winners[-1]
print(winners)
```

```python
#We can use del with a key inside a dictionary to remove unwanted key:value pairs from the dictionary.
product = {'category': 'book',
           'price': 4.99,
           'in_shop': True}

del product['in_shop']
print(product)
```

<p style="text-align:center;">Slice Notation</p>

Sometimes we want to retrieve multiple values from a list. We can do this using slicing. Let's look at the [start:stop] slicing syntax.

```python
ingredients = ["eggs", "flou", "sugar", "salt"]
print(ingredients[0:4])
```

```python
# By coding [3:], we're retrieving all values beginning at
# list position 3 to the end of the list.

letters = ["A", "B", "C", "D", "E"]
print(letters[3:])
print(letters[1:])
```

```python
grades = ['A', 'B', 'C', 'D', 'E']
print(grades[:3])
```

```python
# We can also use a format with two colons, [start:stop:step] , where step
# determines how Python steps between start and end.
# steps like skip

alphabet = ["A", "B", "C", "D", "E", "F"]

print(alphabet[1:6:2])
```
