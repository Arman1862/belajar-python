```python
# Memisahkan string menggunakan split method
```

```python
name = "Arman Rafardhan"

split_name = name.split()
print(split_name)
```

```python
# kalo ada pemisah spesifik kayak "," ";" masukin ke argument nya

name = "Pangestu;Ramadhan"

split_name = name.split(";")
print(split_name)
```
