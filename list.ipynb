Array Maybe

```python
buahs = ['Semangka', 'Ikan', 'Pepaya', "Mangga"]
```

```python
# Array di python disebut list
counter = 0
print("List Buah Buahan: ")
for buah in buahs:
    counter += 1
    print(f"No {counter}. {buah}")
```

manipulasi angka

```python
digit = [20, 19, 18]
```

```python
# ada min, max, sum bla bla bla
print(f"Total nilai array {len(digit)}")
print(f"Ini yang terbesar: {max(digit)}")
print(f"Ini yang terkecil: {min(digit)}")
print(f"Ini yang totalnya: {sum(digit)}")
```

```python
# Sortir Angka (dari yang terkecil)
print(f"kalo gak disortir: {digit}")
print(f"kalo disortir: {sorted(digit)}")
# Atau
digit.sort()
print(f"kalo disortir: {digit}")
```

```python
digit_sort = sorted(digit)
for sortir in digit_sort:
    print(sortir)
```

```python
# Sortir Huruf (berdasarkan alphabet)
buahs.sort()
print(buahs)
```
